#### 연습문제 3. 배열 이어붙이기

- x : [1, 2, 3]

#### 연습문제 4. 브로드 캐스팅
아래 10x3 배열을 중앙 정렬하세요. (X - X_mean)
- X : (10x3) shape을 가진 matrix
- X_mean : 각 열의 평균값 (3, )

```python
X = np.arange(30).reshape((10,3))
X
```

```python
X.shape
```

```python
X_mean = X.mean(axis=1)
```

```python
X_mean
```

```python
X - X_mean
```
